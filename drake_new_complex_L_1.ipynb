Further analysis of generalized cosmic lonelines equation ($L = 1$ and $L$ > 10000 years).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Parameters
d = 1.0  # Density
p_values = [0.01, 0.1, 0.5]  # Different probabilities of finding life
L = 1.0  # Fixed length scale for curve comparison

# Dimensions
n_values = np.arange(1, 21)

# Drake generalization function
def drake(n, L, d, p):
    return (np.pi ** (n / 2) / gamma(n / 2 + 1)) * (L ** n) * d * p

# Plot multiple p curves
plt.figure(figsize=(10, 6))
for p in p_values:
    plt.plot(n_values, [drake(n, L, d, p) for n in n_values], marker='o', label=f'$p$ = {p}')
plt.title("Generalized Cosmic Loneliness Equation vs Dimensions for Various $p$")
plt.xlabel("Number of Dimensions ($n$)")
plt.ylabel("Expected Sites")
plt.legend()
plt.grid(True)
plt.xticks(n_values)
plt.show()



# Heatmap: value vs n and L
L_values = np.linspace(0.5, 2.0, 50)
heatmap = np.array([[drake(n, L_val, d, 0.1) for n in n_values] for L_val in L_values])

plt.figure(figsize=(10, 6))
plt.imshow(heatmap, aspect='auto', origin='lower', extent=[1, 20, 0.5, 2.0])
plt.colorbar(label='Expected Sites')
plt.title("Heatmap of Expected Sites vs Dimension ($n$) and Lifespan of civilization ($L$)")
plt.xlabel("Number of Dimensions ($n$)")
plt.ylabel("Lifespan of civilization ($L$)")
plt.show()
```

Multiple $p$ Curves Plotted the expected number of life‑bearing sites for $p=0.01, 0.1, 0.5$ (with $L=1, d=1$). You can see how the overall scale of the curve scales linearly with $p$, while the peak dimension remains the same.

Heatmap over $n$ and $L$ Held $p=0.1, d=1$, and varied the length scale $L$ from 0.5 to 2.0. This shows how both raising the dimensionality and changing the search radius jointly affect the expected count.
