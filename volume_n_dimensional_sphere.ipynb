# Introduction
A simulation to show how the volume of an $n$-dimensional sphere changes as you change $n$.

## Define a function to calculate the volume of an n-dimensional sphere

### Subtask:
Implement a function that takes the dimension `n` and radius `r` as input and returns the volume of the n-dimensional sphere.


**Reasoning**:
Implement the `sphere_volume` function as described in the instructions.



```python
import math
from scipy.special import gamma
```

```python

def sphere_volume(n, r):
  """Calculates the volume of an n-dimensional sphere.

  Args:
    n: The dimension of the sphere.
    r: The radius of the sphere.

  Returns:
    The volume of the n-dimensional sphere.
  """
  volume = (math.pi**(n/2) / gamma(n/2 + 1)) * (r**n)
  return volume
```

## Generate data

### Subtask:
Generate data by creating a range of values for `n` (e.g., from 1 to 10) and calculate the corresponding volumes for a fixed radius (e.g., radius = 1) using the function defined in the previous step.


**Reasoning**:
Generate a range of n values, set a fixed radius, calculate the volume for each n using the previously defined function, and store the results.



```python
n_values = range(1, 11)
radius = 1
volumes = [sphere_volume(n, radius) for n in n_values]
print("n values:", list(n_values))
print("Volumes:", volumes)
```

## Visualize the results

### Subtask:
Visualize the calculated volumes against the corresponding values of `n` to show how the volume changes as the dimension increases.


**Reasoning**:
Generate a plot to visualize the relationship between the dimension and the volume of the sphere.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(n_values, volumes)
plt.xlabel("Dimension ($n$)")
plt.ylabel("Volume of $n$-dimensional Sphere")
plt.title(f"Volume of $n$-dimensional Sphere as a Function of Dimension (radius={radius})")
plt.grid(True)
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   The volume of an n-dimensional sphere with a radius of 1 was calculated for dimensions from 1 to 10.
*   The calculated volumes show that as the dimension `n` increases, the volume initially increases, reaches a peak around n=5, and then starts to decrease.
