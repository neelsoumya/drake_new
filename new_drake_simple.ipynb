```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Parameters
d = 1.0  # Density
p_values = [0.01, 0.1, 0.5]  # Different probabilities of finding life
L = 10000.0  # Updated length scale

# Dimensions
n_values = np.arange(1, 21)

# Drake generalization function
def drake(n, L, d, p):
    return (np.pi ** (n / 2) / gamma(n / 2 + 1)) * (L ** n) * d * p

# Plot multiple p curves with updated L
plt.figure(figsize=(10, 6))
for p in p_values:
    plt.plot(n_values, [drake(n, L, d, p) for n in n_values], marker='o', label=f'p = {p}')
plt.title("Generalized Drake Equation vs Dimensions\nfor Various $p$ ($L$ = 10,000)")
plt.xlabel("Number of Dimensions ($n$)")
plt.ylabel("Expected Number of Life-Bearing Sites")
plt.legend()
plt.grid(True)
plt.xticks(n_values)
plt.savefig('drake_new_simple_L_10000.png')
plt.show()
plt.close()
```
